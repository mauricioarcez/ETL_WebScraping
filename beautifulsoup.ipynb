```python
# importamos las librerias ya instaladas
from bs4 import BeautifulSoup
import requests
```

### EXTRACT

```python
# obtener el HTML
URL_BASE = 'https://scrapepark.org/spanish/'
consulta = requests.get(URL_BASE)
html_obtenido = consulta.text

# parsear el HTML
soup = BeautifulSoup(html_obtenido, "html.parser")
soup
```

### TRANSFORM

     metodo .find()

```python
#busca y selecciona el primer metodo encontrado
primer_h2 = soup.find("h2") 
print(primer_h2)

#para seleccionar solo texto .text
print(primer_h2.text)
#otra opcion valida
print(soup.h2.text)
```

    .findall()
    .text
    .get_text(strip = TRUE)

```python
#devuelve una lista con todos los resultados de etiquetas
todos_h2 = soup.find_all("h2")
print("\n1.", todos_h2)

#usando limit= podemos limitar el tamaño de la lista al deseado 
un_h2 = soup.find_all("h2", limit=1)
print("\n2." , un_h2)

#esto permite iterar cada seccion y personalizarlas para por ejemplo mostrar solo texto
for seccion in todos_h2:
    print(seccion.text)

#una version mas avanzada de .text es .get_text que permite modificar el texto
#como por ejemplo quitar tabulaciones y espacios en blancos con strip=True
print("\n3.")
for seccion in todos_h2:
    print(seccion.get_text(strip=True))
```

    seleccion de atributos y clases 

```python
# la palabra class esta reservada por ende se debe utilizar con un guion bajo al final
# se puede seleccionar div que adentro tengan clases o solamente la clase. 
# debe ser importante mirar la web scrapeada y verificar si todas las secciones de la web contienen la clase o no
divs = soup.find_all("div", class_ = "heading-container heading-center")
#imprime la lista con todos los divs 
print("\n1.",divs) 

print("\n2.")
for div in divs:
    print(div.get_text(strip=True)) #
```

    descargar imagenes

```python
#seleccionar todas las etiquetas src con True
todo_src = soup.find_all(src=True)
print("\n1.",todo_src)

#al llamar a todos los src tambien estamos llamando a las etiquetas con scripts, png, logos, svg, alts por eso seleccionaremos las etiquetas de solo imagenes
print("\n2.")
for elemento in todo_src:
    if elemento["src"].endswith(".jpg"):
        print(elemento)
```

```python
#conector de archivos y carpetas 
import os 

# lista donde se guardaran las imagenes 
url_imagenes = []

# i es el indice, imagen es la imagen iterada de todo_src 
for i , imagen in enumerate(todo_src):
    if imagen["src"].endswith("png"):
        print(imagen["src"])
        #llama a la web (url_base), en la seccion donde se encuentra la imagen (imagen["src"]) y la trae al programa
        r = requests.get(f'{URL_BASE}{imagen["src"]}') 
        
        #verificar estado de la solicitud
        if r.status_code == 200:
            img = f'imagen_{i}.png'
            #la carpeta debe estar ya creada para que el codigo no se rompa
            ruta = os.path.join("carpeta_img", img)
            #llama a la ruta de la imagenes y lo abre en modo de escritura ('wb') para escribirle la imagen con el nombre (f)
            with open(ruta, 'wb') as f:
                # .write es para escribir la imagen que contendra y .content es para poner el contenido del get()
                f.write(r.content)
                
        #manejar la excepcion en caso de error de estado de solicitud que no sea 200
        else:
            print("No se pudo descargar la imagen")
```

    iframes y tablas


```python
#los iframes no pueden encontrarse dentro del html al filtrarlos con .find_all() 
# ya que son tablas que crea una nueva pagina dentro de mi pagina
# y se codifica dentro del src con table.html
# [0] es para que solo tome el primer valor de la lista. ya que si no mostrara todos los encontrados
URL_TABLA = soup.find_all("iframe")[0]['src']

requests_tabla = requests.get(f'{URL_BASE}{URL_TABLA}')

#volver a hacer el mismo proceso que con el parser html comun pero para las tablas
html_tabla = requests_tabla.text
soup_tabla = BeautifulSoup(html_tabla, "html.parser")
soup_tabla.find_all('table')

#filtrar busqueda por estilos con attrs. en este caso los productos faltantes tienen color rojo en la web
# el html debe estar completo. con sus comillas y ; 
#esto llamara a todo el html
productos_faltantes = soup_tabla.find_all(["th","td"], attrs={"style": "color: red;"})
#por eso con este comprehension list llamo solo al texto 
productos_faltantes = [talle.text for talle in productos_faltantes]

if requests_tabla.status_code == 200:
    print(productos_faltantes)
else:
    print("Error de requests")
```

    extraer datos de un producto

```python
#crear la lista para luego iterarla
divs = soup.find_all("div", class_ = "detail-box")
#crear una lista vacia donde se guardaran los datos extraidos 
productos = []
precios = []

for div in divs:
    if (div.h5 is not None) and ("Patineta" in div.h5.text) and (div.h6 is not None):
        producto = div.h5.get_text(strip=True)
        #reeemplazar el signo pesos para que quede solo numeros y sea tipo numerico
        precio = div.h6.get_text(strip=True).replace("$", "") 
        
        # :<16 es para que quede todo nivelado en un mismo renglon. es la cntidad de caracteres de espacio que deja cada variable
        print(f'{producto:<16} | ${precio}')
        productos.append(producto)
        precios.append(precio)
                
```

```python
#esta pagina contiene 2 contactos por ende la url cambia para cada contacto. entonces.. 
URL_BASE = "https://scrapepark.org/spanish/contact"

#crear un bucle que itere a cada url para cada contacto
# va de 1 a 2 ya que el ultimo numero del range no lo toma
for n in range(1,3):
    URL_FINAL = f'{URL_BASE}{n}'
    
    r = requests.get(URL_FINAL).text
    soup = BeautifulSoup(r, 'html.parser')
    print(soup.h5.text)
```

    datos que no sabemos en que parte de la pagina se encuentran

```python
#Expresiones regulares
import re

#Obtener el html
URL_BASE = "https://scrapepark.org/spanish/"
r = requests.get(URL_BASE)
html_obtenido = r.text

#parsear el html
soup = BeautifulSoup(html_obtenido, "html.parser")

#supongamos que queremos buscar el telofono 4-444-444 o el mail pero no lo encontramos. podemos encontrarlo con r''
telefonos = soup.find_all(string=re.compile(r"\d+-\d+-\d+"))
print(telefonos)
emails = soup.find_all(string=re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'))
print(emails)
```

    moviendonos por el arbol (padre/hijo)


```python
#supongamos que queremos llamar al copyright de la web
copyright = soup.find_all(string=re.compile("©"))
# aparece solo el año ya que lo demas esta escrito por fuera del <p>. en un <a>, un <span>
print(copyright)

#entonces debe llamar al arbol padre con .parent 
#para que funcione debe ser llamado al primer copy. con findall busca todos
#esto tambien sirve para saber donde o en que etiqueta esta posicionado ya que llamara a todas las etiquetas padre 
primer_copyright = copyright[0]
print(primer_copyright.parent)

#tambien podemos llamar a los hermanos o etiquetas del mismo nivel
primer_copyright.parent.find_next_sibling()

```

    control de excepciones

```python
#ademas de los codigos de estado, el control de excepciones try-except puede ser util
# para que el programa nunca deje de correr ni se frene
# es sensible ante mayusculas y tildes asi que la busqueda debe ser perfecta 
string_a_buscar = ['MENÚ', 'MENU','Patineta', 'menu', 'mauricio']

for string in string_a_buscar:
    #en caso de no usar try-except saldra atributteError y se frena el programa
    try:
        busqueda = soup.find(string=re.compile(string))
        print(busqueda.parent)
    except AttributeError:
        print(f'no se pudo encontrar la palabra: {string}')
                             
```

```python
#con la ya lista creada de precios y productos anteriormente 
#le insertamos un string en el indice 0 de la lista que se llame como su variable
productos.insert(0, "productos")
precios.insert(0, "precios")

#para luego transformarla en un diccionario conjunto 
datos = dict(zip(productos, precios))
datos
```
